```python
weather = weather.drop(['Date','RISK_MM'], axis = 1)
```

```python
weather.info()
```

```python
#Drop rows where column value is 'NA'
weather = weather.dropna(subset = ['RainToday'])
```

```python
weather.fillna(0, inplace = True)
weather.head()
```

```python
weather.isnull()
```

```python
weather['WindGustDir'] = weather["WindGustDir"].apply(Directions)
weather['WindDir9am'] = weather['WindDir9am'].apply(Directions)
weather['WindDir3pm'] = weather['WindDir3pm'].apply(Directions)
```

```python
#Get unique values
weather.WindDir3pm.unique()
```

```python
#Get Dummies -Convert Categorical Data into Numerical Data
weather = pd.get_dummies(weather, columns = ['RainToday'])
weather = pd.get_dummies(weather, columns = ['WindDir3pm'])
```

```python
weather = pd.get_dummies(weather, columns = ['WindDir9am', 'WindGustDir'])
```

```python
weather = pd.get_dummies(weather, columns = ['Location'])
```

```python
def Rain_Tomorrow(x):
    if x == 'Yes':
        return 1
    if x == 'No':
        return 0

weather['RainTomorrow'] = weather['RainTomorrow'].apply(Rain_Tomorrow)
```

```python
#Data Exploration
weather['RainTomorrow'].value_counts()
```

```python
sns.countplot(x = 'RainTomorrow', data = weather, palette = 'hls')
plt.show()
```

```python
#Event Rate or Percentage's
count_rainTomorrow = len(weather[weather['RainTomorrow'] == 1])
count_notRainTomorrow = len(weather[weather['RainTomorrow'] == 0])
percent_rainTomorrow = count_rainTomorrow / (count_rainTomorrow + count_notRainTomorrow)
percent_notRainTomorrow = count_notRainTomorrow / (count_rainTomorrow + count_notRainTomorrow)
print("Percentage of Rain Tomorrow:", percent_rainTomorrow*100)
print("Percentage of not Raining Tomorrow:", percent_notRainTomorrow*100)
```

```python
#Logistic Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x = weather.drop(['RainTomorrow'], axis = 1)
y = weather.RainTomorrow

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
#Create an instance and fit the model
logit_model = LogisticRegression()
logit_model.fit(x_train, y_train)
```

```python
#Predictions
Predictions = logit_model.predict(x_test)
```

```python
#Model Evaluation
from sklearn.metrics import classification_report

print(classification_report(y_test, Predictions))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, Predictions))
```
